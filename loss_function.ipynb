```python
hours_studied = [1, 2, 3, 4, 5]
actual_scores = [20, 40, 60, 80, 100]

# We are taking Example of that model Prdicxted Something

predicted_scores = [50,50,50,50,50] 
```

```python
# calculating the error

errors = [] # Creating the Empty 
for i in range(len(actual_scores)):
    error = actual_scores[i]-predicted_scores[i]
    errors.append(error)

print("Errors:", errors)

# Mean Absolutre Error (onlyh Sign Are Ignored)

mae = sum([abs(e)for e in errors]) / len(errors)
print(mae)
# MSE (Scaled the Error using the Squre)

mse = sum([e**2 for e in errors]) / len(errors)
print(mse)
```
